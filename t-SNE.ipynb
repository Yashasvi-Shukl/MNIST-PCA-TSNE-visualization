# Visualization using t-SNE

```python
# Choosing top 15000 points
sample1 = data.head(15000)
value1 = label.head(15000)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
new_sample1 = sc.fit_transform(sample1)
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components = 2, random_state = 0, perplexity = 50, n_iter = 6000 )
tsne_data = model.fit_transform(sample1)
```

```python
# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T, value1)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))

# Ploting the result of tsne
sn.FacetGrid(tsne_df, hue="label", size=10).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

```python
# Running t-SNE multiple times with different values
i = 0
perplexity = [50, 100]
iter_ = [5000, 8000]

for val in (25, 50):
    sample2 = data.head(val)
    value2 = label.head(val)


    sc = StandardScaler()
    new_sample1 = sc.fit_transform(sample1)


    model = TSNE(n_components = 2, random_state = 0, perplexity = perplexity[i], n_iter = iter_[i])
    tsne_data = model.fit_transform(sample1)

    # creating a new data frame which help us in ploting the result data
    tsne_data = np.vstack((tsne_data.T, value1)).T
    tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
    i += 1
    # Ploting the result of tsne
    sn.FacetGrid(tsne_df, hue="label", size=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
    plt.show()
```
